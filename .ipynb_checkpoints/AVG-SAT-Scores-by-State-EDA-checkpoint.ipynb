```python
dfdrug = pd.read_csv('drug-use-by-age.csv')
dfdrug.info()

# Our data does require some cleaning
# We have no null values but some of our coulmns need to be converted from object type to float type
```

remove '-' from frequency column for each drug and replace with np.nan as a place holder

```python
dfdrug['cocaine-frequency'].replace('-', np.nan, inplace=True)
dfdrug['cocaine-frequency'] = dfdrug['cocaine-frequency'].apply(lambda x: float(x))
dfdrug['crack-frequency'].replace('-', np.nan, inplace=True)
dfdrug['crack-frequency'] = dfdrug['crack-frequency'].apply(lambda x: float(x))
dfdrug['heroin-frequency'].replace('-', np.nan, inplace=True)
dfdrug['heroin-frequency'] = dfdrug['heroin-frequency'].apply(lambda x: float(x))
dfdrug['inhalant-frequency'].replace('-', np.nan, inplace=True)
dfdrug['inhalant-frequency'] = dfdrug['inhalant-frequency'].apply(lambda x: float(x))
dfdrug['oxycontin-frequency'].replace('-', np.nan, inplace=True)
dfdrug['oxycontin-frequency'] = dfdrug['oxycontin-frequency'].apply(lambda x: float(x))
dfdrug['meth-frequency'].replace('-', np.nan, inplace=True)
dfdrug['meth-frequency'] = dfdrug['meth-frequency'].apply(lambda x: float(x))
```
